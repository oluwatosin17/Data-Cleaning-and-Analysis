### Apply a Function Element-Wise to Multiple Columns Using Applymap Method

```python
def label(element):
    if element > 1:
        return 'High'
    else:
        return 'Low'
economy_apply = happiness2015['Economy'].apply(label)
factors = ['Economy','Family','Health','Freedom','Trust','Generosity']
factor_impact = happiness2015[factors].applymap(label)
```

```python
factor_impact.head()
```

```python
factor_impact.apply(pd.value_counts)
```

```python
def v_counts(col):
    num = col.value_counts()
    den = col.size
    return num/den
    
```

```python
factor_impact.apply(v_counts)
```

The melt functions allows the values of economy, health, and family to reside in the same column

```python
main_cols = ['Country','Region','Happiness Rank','Happiness Score']
factors = ['Economy','Family','Health','Freedom','Trust','Generosity','Dystopia Residual']
```

```python
melt = pd.melt(happiness2015, id_vars = main_cols, value_vars = factors)
percentages = (melt['value']/melt['Happiness Score'])*100
percentages = round(percentages,2)
melt['Percentage'] = percentages
```

```python
melt.head()
```

```python
pv_melt = melt.pivot_table(index = 'variable', values = 'value')
```

```python
pv_melt.plot(kind = 'pie',y = 'value',legend = False)
```
