Specify the join type in the `how`  command. A left join or left merge keeps every row from the left dataframe. and whats is common.

Different column names are specified for merges in Pandas using the `left_on` and `right on` parameters, instead of using only the`on` parameter.

### Extracting Substrings from a Series

Suppose we wanted to extract any year mentioned in the specialnotes 

1.  Pattern =  r"[0-9]" 
2.  lower case = r"[a-z]" same as upper case

if we wanted to search for a three character substring in a column that starts with the 1 and ends with 6 with two letters of any kind

pattern = r"[1-6][a-z][a-z]" = r"[1-6][a-z]{2}"

```python

pattern = r"([1-2][0-9][0-9][0-9])"

years = merged['SpecialNotes'].str.extract(pattern)
```

```python

pattern = r"([1-2][0-9][0-9][0-9])"

years = merged['SpecialNotes'].str.extract(pattern, expand = True)
```

### Extracting All Matches of a Pattern from a series

```python
merged = merged.set_index('Country')
```

Using a named capturing group means we can refer to that group by a specific name instead of just a number we can use

(?P<Column_Name>...)

```python
pattern = r"(?P<Years>[1-2][0-9]{3})"
merged['SpecialNotes'].str.extractall(pattern)
```

```python
years = merged['IESurvey'].str.extractall(pattern)
value_counts = years['Years'].value_counts()
```

```python
value_counts
```

### Extracting more than one group of patterns from a series

Working on something like 2018/19

```python
pattern = r"(?P<First_Year>[1-2][0-9]{3})(/)?(?P<Second_Year>[0-9]{2})"
years = merged['IESurvey'].str.extractall(pattern)
```

```python
years
```

```python
years.sort_values('Second_Year')
```

```python
first_two_year = years['First_Year'].str[0:2]
years['Second_Year'] = first_two_year+years['Second_Year']
```
